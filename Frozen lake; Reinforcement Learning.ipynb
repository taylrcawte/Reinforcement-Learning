# Reinforcement Learning Example; The Frozen Lake

In this example a reinforcement learning agent is taught to navigate a "frozen lake" through the use of q learning. The example makes use of the 'gym' library; a useful reinforcement learning environment library. 

```python
# library imports 
import gym
import numpy as np
import matplotlib.pyplot as plt
import time, pickle, os
from statistics import mean
%matplotlib inline
```

The below helper functions allow for the learning agent to choose an action and learn which actions are most likely to help the agent succeed in crossing the like. Actions taken to successfully cross the lake are rewarded while actions which lead to an unsuccesful crossing are not. A more detailed description of the q learning process is provided in the repository. 

```python
def choose_action(state):
    action=0
    if np.random.uniform(0, 1) < epsilon:
        action = env.action_space.sample()
    else:
        action = np.argmax(Q[state, :])
    return action

def learn(state, state2, reward, action):
    predict = Q[state, action]
    target = reward + gamma * np.max(Q[state2, :])
    Q[state, action] = Q[state, action] + lr_rate * (target - predict)
```

The environment is now initialized and a some initial parameters are defined. 

```python
env = gym.make('FrozenLake-v0')

epsilon = 0.9
total_episodes = 2000
max_steps = 100

lr_rate = 0.81  
gamma = 0.96

Q = np.zeros((env.observation_space.n, env.action_space.n))
scores = []
```

```python
for episode in range(total_episodes):
    state = env.reset()
    t = 0

    while t < max_steps:
        env.render()

        action = choose_action(state)  

        state2, reward, done, info = env.step(action)  

        learn(state, state2, reward, action)

        state = state2

        t += 1

        print(episode)

        if done:
            break

        time.sleep(0.1)
 
    score = np.round(np.sum(Q/np.max(Q)*100))
    scores.append(score)
    
print(Q)
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('Episod')
ax.set_ylabel('Score')
ax.plot(scores)
ax.tick_params(axis='y')
fig.set_size_inches(10, 7)
fig.savefig('Convergence.png', dpi=100)
```

# Hyperparameter Tuning
The hyperparameters are: Epsilon, Gamma and alpha (Learning Rate). For this type of algorithm, tuning is done by passing values to the hyperparameter and selecting the most ideal which is the minimum of (average) time that is reqired for the agent to reach the final point or destination. 


## Epsilon Tuning

For the hyperparameter tuning of epsilon, we provide a list of possible values to begin the epsilon greedy policy.

```python
epsilonList = [0.01, 0.1, 0.2, 0.3,0.5, 0.7, 0.9]
mean_timesteps = []

for e in epsilonList:

    env = gym.make('FrozenLake-v0')
    epsilon = e
    total_episodes = 400
    max_steps = 100
    lr_rate = 0.81  
    gamma = 0.96

    Q = np.zeros((env.observation_space.n, env.action_space.n))
    scores = []
    timesteps = []
    for episode in range(total_episodes):
        state = env.reset()
        t = 0

        while t < max_steps:

            action = choose_action(state)  

            state2, reward, done, info = env.step(action)  

            learn(state, state2, reward, action)

            state = state2

            t += 1

            if done:
                timesteps.append(t)
                break

            time.sleep(0.1)

        print("Episod number:", episode, "Epsilon:", e)
        print(timesteps)
        score = np.round(np.sum(Q/np.max(Q)*100))
        scores.append(score)

    
    mean_timesteps.append(mean(timesteps))
    fig, ax = plt.subplots()
    ax.set_xlabel('Episod')
    ax.set_ylabel('Score')
    ax.plot(scores, label = "Epsilon = " + str(e))
    ax.tick_params(axis='y')
    ax.legend()
    fig.set_size_inches(10, 7)
    
fig.savefig('Epsilon_Convergence.png', dpi=100)
```

```python
fig, ax = plt.subplots()
color = 'red'
ax.set_xlabel('Epsilon: Exploration rate')
ax.set_ylabel('Mean total time')
ax.plot(epsilonList, mean_timesteps, color=color)
ax.tick_params(axis='y')
fig.set_size_inches(10, 7)
fig.savefig('Epsilon.png', dpi=100)
```

## Gamma Tuning

```python
gammaList = [0.01, 0.05, 0.1, 0.2, 0.3, 0.4, 0.6, 0.8]
mean_timesteps = []

for g in gammaList:

    env = gym.make('FrozenLake-v0')
    epsilon = 0.9
    total_episodes = 400
    max_steps = 100
    lr_rate = 0.81  
    gamma = g

    Q = np.zeros((env.observation_space.n, env.action_space.n))
    scores = []
    timesteps = []
    for episode in range(total_episodes):
        state = env.reset()
        t = 0

        while t < max_steps:

            action = choose_action(state)  

            state2, reward, done, info = env.step(action)  

            learn(state, state2, reward, action)

            state = state2

            t += 1

            if done:
                timesteps.append(t)
                break

            time.sleep(0.1)
            
        print("Episod number:", episode, "Gamma:", g)
        print(timesteps)
        score = np.round(np.sum(Q/np.max(Q)*100))
        scores.append(score)

    
    mean_timesteps.append(mean(timesteps))
    fig, ax = plt.subplots()
    ax.set_xlabel('Episod')
    ax.set_ylabel('Score')
    ax.plot(scores, label = "gamma = " + str(g))
    ax.tick_params(axis='y')
    ax.legend()
    fig.set_size_inches(10, 7)
    fig.savefig('Gamma_Convergence.png', dpi=100)
```

### Learning Rate Tuning

For the hyperparameter tuning of Gamma, we provide a list of possible values.

```python
lr_rateList = [0.1, 0.2, 0.3, 0.4, 0.6, 0.8, 0.9, 0.95]
mean_timesteps = []

for lr in lr_rateList:

    env = gym.make('FrozenLake-v0')
    epsilon = 0.9
    total_episodes = 400
    max_steps = 100
    lr_rate = lr  
    gamma = 0.3

    Q = np.zeros((env.observation_space.n, env.action_space.n))
    scores = []
    timesteps = []
    for episode in range(total_episodes):
        state = env.reset()
        t = 0

        while t < max_steps:

            action = choose_action(state)  

            state2, reward, done, info = env.step(action)  

            learn(state, state2, reward, action)

            state = state2

            t += 1

            if done:
                timesteps.append(t)
                break

            time.sleep(0.1)
            
        print("Episod number:", episode, "Learning Rate:", lr)
        print(timesteps)
        score = np.round(np.sum(Q/np.max(Q)*100))
        scores.append(score)

    
    mean_timesteps.append(mean(timesteps))
    fig, ax = plt.subplots()
    ax.set_xlabel('Episod')
    ax.set_ylabel('Score')
    ax.plot(scores, label = "lr_rate = " + str(lr))
    ax.tick_params(axis='y')
    ax.legend()
    fig.set_size_inches(10, 7)
    fig.savefig('LR_Convergence.png', dpi=100)
```
